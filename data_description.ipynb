```python
#ich ben√∂tige noch den Median (median), den Modus (mode) sowie die Varianz (variance).
import pandas as pd
avocado = pd.read_csv("avocado.csv")
print ("Statistiken Spalte 'TotalBags'")
mode_Total_bags = avocado["TotalBags"].mode()
print("Modus: ", mode_Total_bags)

median_Total_bags = avocado["TotalBags"].median()
print("Median: ", median_Total_bags)

variance_Total_bags = avocado["TotalBags"].var()
print("Variance: ", variance_Total_bags)
#-------------------------------------------------------------------------#
print("#-------------------------------------------------------------------------#")
print ("Statistiken Spalte 'Average Price'")
mode_avg_price = avocado["AveragePrice"].mode()
print("Modus: ", mode_avg_price)

median_avg_price = avocado["AveragePrice"].median()
print("Median: ", median_avg_price)

variance_avg_price = avocado["AveragePrice"].var()
print("Variance: ", variance_avg_price)
#-------------------------------------------------------------------------#
print("#-------------------------------------------------------------------------#")
print ("Statistiken Spalte 'TotalVolume'")
mode_TV = avocado["TotalVolume"].mode()
print("Modus: ", mode_TV)

median_TV = avocado["TotalVolume"].median()
print("Median: ", median_TV)

variance_TV = avocado["TotalVolume"].var()
print("Variance: ", variance_TV)
#-------------------------------------------------------------------------#
print("#-------------------------------------------------------------------------#")
print ("Statistiken Spalte '4225'")
mode_4225 = avocado["4225"].mode()
print("Modus: ", mode_4225)

median_4225 = avocado["4225"].median()
print("Median: ", median_4225)

variance_4225 = avocado["4225"].var()
print("Variance: ", variance_4225)
#-------------------------------------------------------------------------#
print("#-------------------------------------------------------------------------#")
print ("Statistiken Spalte '4046'")
mode_4046 = avocado["4046"].mode()
print("Modus: ", mode_4046)

median_4046 = avocado["4046"].median()
print("Median: ", median_4046)

variance_4046 = avocado["4046"].var()
print("Variance: ", variance_4046)
#-------------------------------------------------------------------------#
print("#-------------------------------------------------------------------------#")
print ("Statistiken Spalte '4770'")
mode_4770 = avocado["4770"].mode()
print("Modus: ", mode_4770)

median_4770 = avocado["4770"].median()
print("Median: ", median_4770)

variance_4770 = avocado["4770"].var()
print("Variance: ", variance_4770)
#-------------------------------------------------------------------------#
print("#-------------------------------------------------------------------------#")
print ("Statistiken Spalte 'SmallBags'")
mode_SB = avocado["SmallBags"].mode()
print("Modus: ", mode_SB)

median_SB = avocado["SmallBags"].median()
print("Median: ", median_SB)

variance_SB = avocado["SmallBags"].var()
print("Variance: ", variance_SB)
#-------------------------------------------------------------------------#
print("#-------------------------------------------------------------------------#")
print ("Statistiken Spalte 'LargeBags'")
mode_LB = avocado["LargeBags"].mode()
print("Modus: ", mode_LB)

median_LB = avocado["LargeBags"].median()
print("Median: ", median_LB)

variance_LB = avocado["Large Bags"].var()
print("Variance: ", variance_LB)
#-------------------------------------------------------------------------#
print("#-------------------------------------------------------------------------#")
print ("Statistiken Spalte 'XLargeBags'")
mode_XLB = avocado["XLargeBags"].mode()
print("Modus: ", mode_XLB)

median_XLB = avocado["XLargeBags"].median()
print("Median: ", median_XLB)

variance_XLB = avocado["XLargeBags"].var()
print("Variance: ", variance_XLB)
```
